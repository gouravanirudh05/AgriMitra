```python
# Run in Kaggle / Colab / local (adjust input/output paths accordingly)

import os
import shutil
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, Model
import tensorflow_hub as hub
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# -----------------------
# 1) Paths & settings
# -----------------------
# Change BASE_SRC to where the original PlantVillage dataset is mounted
BASE_SRC = tushar5harma_plant_village_dataset_updated_path  # <- adjust if needed
WORK_DIR = "plant_id_dataset"             # new dataset we will create
os.makedirs(WORK_DIR, exist_ok=True)

IMG_SIZE = (224, 224)
BATCH_SIZE = 64
EPOCHS = 8   # start small, increase if you have time & GPU
RND = 42
random.seed(RND)
np.random.seed(RND)
tf.random.set_seed(RND)

# -----------------------
# 2) Collect per-plant images into class folders
# -----------------------
# We aggregate all Train/Val/Test images under each plant into new Train/Val/Test plant folders.
splits = ["Train", "Val", "Test"]

# Make target dirs
for s in splits:
    for plant in os.listdir(BASE_SRC):
        plant_dir = os.path.join(BASE_SRC, plant)
        if not os.path.isdir(plant_dir):
            continue
        tgt = os.path.join(WORK_DIR, s, plant)
        os.makedirs(tgt, exist_ok=True)

# Copy images (lightweight: we copy file paths; for very large datasets you may prefer symlinks)
for plant in os.listdir(BASE_SRC):
    plant_dir = os.path.join(BASE_SRC, plant)
    if not os.path.isdir(plant_dir):
        continue
    for split in splits:
        split_src = os.path.join(plant_dir, split)
        if not os.path.isdir(split_src):
            continue
        # iterate disease subfolders
        for disease in os.listdir(split_src):
            disease_dir = os.path.join(split_src, disease)
            if not os.path.isdir(disease_dir):
                continue
            for fname in os.listdir(disease_dir):
                src_f = os.path.join(disease_dir, fname)
                if not os.path.isfile(src_f):
                    continue
                # copy into plant class folder
                dst_f = os.path.join(WORK_DIR, split, plant, f"{plant}__{disease}__{fname}")
                # avoid collisions: prefix with plant and disease
                try:
                    shutil.copy2(src_f, dst_f)
                except Exception as e:
                    # if disk is limited, you could instead create a symlink:
                    # os.symlink(src_f, dst_f)
                    # but Kaggle may restrict symlinks. We'll ignore copy errors.
                    print("copy error:", e, src_f)

print("Created plant-class dataset at:", WORK_DIR)

# -----------------------
# 3) Quick dataset counts / EDA
# -----------------------
def count_images(base):
    counts = {}
    for plant in sorted(os.listdir(base)):
        p = os.path.join(base, plant)
        if os.path.isdir(p):
            counts[plant] = sum(len(files) for _,_,files in os.walk(p))
    return counts

print("Train counts:", count_images(os.path.join(WORK_DIR, "Train")))
print("Val counts:  ", count_images(os.path.join(WORK_DIR, "Val")))
print("Test counts: ", count_images(os.path.join(WORK_DIR, "Test")))

# Optional: plot a few example images per plant
plt.figure(figsize=(14,6))
plants = sorted(os.listdir(os.path.join(WORK_DIR,"Train")))[:8]
i = 1
for plant in plants:
    plant_folder = os.path.join(WORK_DIR, "Train", plant)
    imgs = os.listdir(plant_folder)
    if not imgs:
        continue
    img_path = os.path.join(plant_folder, random.choice(imgs))
    img = plt.imread(img_path)
    plt.subplot(2, 4, i)
    plt.imshow(img.astype('uint8'))
    plt.title(plant)
    plt.axis('off')
    i += 1
plt.tight_layout()
plt.show()

# -----------------------
# 4) Data generators
# -----------------------
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_dir = os.path.join(WORK_DIR, "Train")
val_dir   = os.path.join(WORK_DIR, "Val")
test_dir  = os.path.join(WORK_DIR, "Test")

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)
val_gen = val_datagen.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)
test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

class_indices = train_gen.class_indices
inv_class_map = {v:k for k,v in class_indices.items()}
print("Classes:", class_indices)

# -----------------------
# 5) Build model (reuse your TF-Hub ViT feature extractor)
# -----------------------
# Use same module as your example. It's a feature-extractor; trainable=False for quick convergence.
vit_model = hub.load("https://tfhub.dev/sayakpaul/vit_r50_l32_fe/1")

def vit_features(x):
    return vit_model(x)
def build_vit_model(num_classes, model_name):
    """Build a ViT-based classification model using Functional API."""
    # Define input
    inputs = tf.keras.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3), name="input_image")
    x = layers.Lambda(vit_features, name="vit_feature_extractor")(inputs)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation="softmax", name="output_layer")(x)

    # Build the model
    model = tf.keras.Model(inputs=inputs, outputs=outputs, name=model_name)

    model.compile(
        loss="categorical_crossentropy",
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
        metrics=["accuracy"]
    )

    return model

model = build_vit_model(len(class_indices), "plant_id_model")
model.summary()

# -----------------------
# 6) Train
# -----------------------
callbacks = [
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, verbose=1),
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=4, restore_best_weights=True, verbose=1)
]

history = model.fit(
    train_gen,
    epochs=EPOCHS,
    validation_data=val_gen,
    callbacks=callbacks,
    steps_per_epoch=len(train_gen),
    validation_steps=len(val_gen)
)

# -----------------------
# 7) Evaluate on test set & produce reports
# -----------------------
# Predict on test set
test_steps = len(test_gen)
Y_pred = model.predict(test_gen, steps=test_steps, verbose=1)
y_pred = np.argmax(Y_pred, axis=1)
y_true = test_gen.classes

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=[inv_class_map[i] for i in range(len(inv_class_map))],
            yticklabels=[inv_class_map[i] for i in range(len(inv_class_map))], cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title(f'Confusion Matrix (accuracy={accuracy_score(y_true, y_pred)*100:.2f}%)')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Classification report
print("Classification report:\n")
print(classification_report(y_true, y_pred, target_names=[inv_class_map[i] for i in range(len(inv_class_map))]))

# -----------------------
# 8) Save model
# -----------------------
MODEL_SAVE = "/kaggle/working/plant_identification_model"
model.save(MODEL_SAVE)
print("Saved model to:", MODEL_SAVE)

# Optionally zip it for download:
# !zip -r plant_id_model.zip /kaggle/working/plant_identification_model
```
